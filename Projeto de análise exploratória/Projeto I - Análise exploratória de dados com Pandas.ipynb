## Análise da coluna Custo

Criando a coluna *Custo Total* multiplicando o custo unitário pela quantidade.

```python
vendas['Custo Total'] = vendas['Custo Unitário']*vendas['Quantidade']
```

**Produtos com maiores custos totais**: Os produtos com maiores custos totais são o *Adventure Works Laptop 15.4W M1548 Black*, *Adventure Works Laptop 15.4W M1548 Red* e *Fabrikan Trendsetter 2/3" 17mm x 100 Grey*.

```python
(vendas['Custo Total']).groupby(by=vendas['Produto']).sum().sort_values().plot.barh(figsize=(20,5))
plt.title('Custo total por produto',size=15)
plt.yticks(size=18)
plt.ylabel('Produto',size=18);
```

Produtos por custos totais médios.

```python
(vendas['Custo Total']).groupby(by=vendas['Produto']).mean().sort_values().plot.barh(figsize=(20,5), color='orange')
plt.yticks(size=18)
plt.title('Custo total médio por produto',size=15)
plt.ylabel('Produto',size=18);
```

A loja com maior custo total é a 307.

```python
vendas['Custo Total'].groupby(by=vendas['ID Loja']).sum().sort_values().plot.barh(figsize=(10,2.5), color='orange')
plt.title('Custo total por loja',size=15)
plt.yticks(size=18)
plt.ylabel('',size=10);
```

Mas a loja com maior custo médio é a 199.

```python
vendas['Custo Total'].groupby(by=vendas['ID Loja']).mean().sort_values().plot.barh(figsize=(10,2.5), color='green')
plt.title('Custo total médio por loja',size=15)
plt.yticks(size=18)
plt.ylabel('',size=12);
```

Distribuição e *boxplot* da coluna Custo Total : A distribuição não possui uma distribuição normal e essa coluna possui muitos *outliers* acima do limite superior do *boxplot*.

```python
fig, ax = plt.subplots(1,2, figsize=(20,4))
vendas['Custo Total'].plot.hist(100, ax=ax[0])
vendas['Custo Total'].plot.box(100, ax=ax[1]);
```

## Análise da coluna Preço

Criando a coluna preço total multiplicando o preço unitário pela quantidade vendida.

```python
vendas['Preço Total'] = vendas['Preço Unitário']*vendas['Quantidade']
```

*Fabrikan Trendsetter 2/3" 17mm X100 Grey* é o produto com maior soma de preços.

```python
(vendas['Preço Total']).groupby(by=vendas['Produto']).sum().sort_values().plot.barh(figsize=(20,5))
plt.title('Preço total por produto',size=15)
plt.yticks(size=18)
plt.ylabel('Produto',size=18);
```

*Fabrikan Trendsetter 1/3" 8.5mm X200 Black* é o produto com maior preço médio.

```python
(vendas['Preço Total']).groupby(by=vendas['Produto']).mean().sort_values().plot.barh(figsize=(20,5),color='blue')
plt.title('Preço total médio por produto',size=15)
plt.yticks(size=18)
plt.ylabel('Produto',size=18);
```

A loja com maior média de preços é a 199.

```python
vendas['Preço Total'].groupby(by=vendas['ID Loja']).mean().sort_values().plot.barh(figsize=(10,2.5), 
                                                                                   color='orange', stacked=True)
plt.title('Preço total médio por loja',size=15)
plt.yticks(size=18)
plt.ylabel('',size=10);
```

A marca com maior preço médio é a *Fabrikam*.

```python
vendas['Preço Total'].groupby(by=vendas['Marca']).mean().sort_values().plot.bar(figsize=(20,4), 
                                                                                   color='red', stacked=True)
plt.title('Preço total médio por loja',size=15)
plt.xticks(size=18, rotation='22.5')
plt.xlabel('Marca',size=15);
```

Distribuição e *boxplot* da coluna preço total.

```python
fig, ax = plt.subplots(1,2, figsize=(20,6))
vendas['Preço Total'].plot.hist(100, ax=ax[0], color='b')
vendas['Preço Total'].plot.box(100, ax=ax[1]);
```

Gráfico de área entre preço total, custo total e valor de desconto.

```python
vendas[['Custo Total','Valor Desconto','Preço Total']].plot.area(figsize=(20,10));
```

Abaixo podemos ver os montantes de valores de custo, preço e desconto por loja.

```python
vendas[['Custo Total','Preço Total','Valor Desconto']].groupby(by=vendas['ID Loja']).mean().plot.barh(figsize=(20,7))
plt.title('Preço, Custo e Desconto por Loja',size=15)
plt.yticks(size=18)
plt.ylabel('Produto',size=18);
```

Abaixo podemos ver os montantes de valores de custo, preço e desconto por produto.

```python
vendas[['Custo Total','Preço Total','Valor Desconto']].groupby(by=vendas['Produto']).sum().plot.barh(figsize=(20,10))
plt.title('Preço, Custo e Desconto por produto',size=15)
plt.yticks(size=18)
plt.ylabel('Produto',size=18);
```

## Análise do lucro

Criando a variável lucro.

```python
vendas['Lucro'] = vendas['Preço Total'] - vendas['Custo Total'] - vendas['Valor Desconto']
```

**Qual a loja com maior lucro ?** : Loja 199

```python
vendas['Lucro'].groupby(by=vendas['ID Loja']).sum().sort_values().plot.barh(figsize=(20,4), color='red')
plt.title('Lucro por loja',size=15)
plt.yticks(size=18, rotation='22.5')
plt.ylabel('Loja',size=15);
```

A *Fabrikam* é a resposável pela maior parte do lucro.

```python
vendas['Lucro'].groupby(by=vendas['Marca']).sum().sort_values().plot.barh(figsize=(20,4), color='green')
plt.title('Lucro por Marca',size=15)
plt.yticks(size=18, rotation='22.5')
plt.ylabel('Marca',size=15);
```

A classe *Regular* é a responsável pela maior parte do lucro.

```python
vendas['Lucro'].groupby(by=vendas['Classe']).sum().sort_values().plot.barh(figsize=(20,4), color='blue')
plt.title('Lucro por Classe',size=15)
plt.yticks(size=18, rotation='22.5')
plt.ylabel('Classe',size=15);
```

Lucro por Porduto.

```python
vendas['Lucro'].groupby(by=vendas['Produto']).sum().sort_values().plot.barh(figsize=(20,10), color='orange')
plt.title('Lucro por Porduto',size=20)
plt.yticks(size=18)
plt.ylabel('Produto',size=15);
```

Lucro pela cor do produto.

```python
vendas['Lucro'].groupby(by=vendas['Cor']).sum().sort_values().plot.barh(figsize=(20,4), color='green')
plt.title('Lucro por Cor do produto',size=15)
plt.yticks(size=18, rotation='22.5')
plt.ylabel('Cor',size=18);
```

Lucros do anos de 2008 e 2009.

```python
lucro_2009 = vendas[vendas["Data Venda"].dt.year==2009]
lucro_2008 = vendas[vendas["Data Venda"].dt.year==2008]
lucro_2008.groupby(vendas['Data Venda'].dt.month)['Lucro'].sum().plot(figsize=(20,7))
lucro_2009.groupby(vendas['Data Venda'].dt.month)['Lucro'].sum().plot();
```

Distribuição e *boxplot* do lucro.

```python
fig, ax = plt.subplots(1,2, figsize=(20,6))
vendas['Lucro'].plot.hist(100, ax=ax[0], color='b')
vendas['Lucro'].plot.box(100, ax=ax[1]);
```
